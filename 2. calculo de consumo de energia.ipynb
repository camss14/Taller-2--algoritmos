```python
algorithim para calcular el consumo de energia
1. iniciar
2. definir el valor  "potencias en vatios", "horas al dia", "numero de dias en el mes"
3.convertir el valor de la potencia de vatios a kilovatios
4. Calcular el consumo diario multiplicando la potencia en kilovatios por el tiempo de uso diario en horas
5. Calcular el consumo mensual multiplicando el consumo diario por el número de días en el mes
6. mostrar el consumo final mensual
7. finalizar
```

```python
Pseudocodigo
program calcular_consumo_energia
Strat
// programa para calcular el consumo de energia
// declaracion de variables
declare float potencia_vatios
declare int horas al dia
declare int dias_mes
declare float potencia kilovatios
declare float consumo diario
declare float consumo mensual

// obtencion de variables
potencia vatios = 1500
horas al dia = 3
dias mes = 30

// conversion de potencia a kilovatios
potencia kilovatios = potencia vatios / 1000

// calculo del consumo diario
consumo diario = potencia kilovatios * horas al dia

// calculo del consumo mensual
consumo mensual = consumo diario * dias mes

// mostrar el resultado en kilovatios por hora
display consumo de energus en kilovatios por hora: consumo mensual

end program
```
